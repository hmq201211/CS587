```python
import pandas as pd
import os
from pandas import DataFrame, Series           
import json
import numpy as np
import tensorflow as tf
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
weights={
         'in':tf.Variable(tf.random_normal([1,10])),
         'out':tf.Variable(tf.random_normal([10,1]))
         }
biases={
        'in':tf.Variable(tf.constant(0.1,shape=[10,])),
        'out':tf.Variable(tf.constant(0.1,shape=[1,]))
        }
X=tf.placeholder(tf.float32, [None,5,1])
Y=tf.placeholder(tf.float32, [None,5,1])
train_x,train_y=[],[]
```

```python
def build_data(normalize_data):
    for i in range(len(normalize_data)-5-1):
        x=normalize_data[i:i+5]
        y=normalize_data[i+1:i+5+1]
        train_x.append(x.tolist())
        train_y.append(y.tolist())
```

```python
def build_model(model_degree):      
    weights_in=weights['in']
    biases_in=biases['in']
    input=tf.reshape(X,[-1,1])  
    input_rnn=tf.matmul(input,weights_in)+biases_in
    input_rnn=tf.reshape(input_rnn,[-1,5,10])  
    cell=tf.nn.rnn_cell.BasicLSTMCell(10)
    init_state=cell.zero_state(1,dtype=tf.float32)
    with tf.variable_scope('scope', reuse=tf.AUTO_REUSE):
        output_rnn,final_states=tf.nn.dynamic_rnn(cell, input_rnn,initial_state=init_state, dtype=tf.float32)
    output=tf.reshape(output_rnn,[-1,10]) 
    weight_out=weights['out']
    biases_out=biases['out']
    prediction=tf.matmul(output,weight_out)+biases_out
    return prediction,final_states
def train_and_predict():
    prediction,final_states=build_model(1)      
    saver=tf.train.Saver(tf.global_variables())
    with tf.Session() as session:
        session.run(tf.global_variables_initializer())
        previous_sequence=train_x[-1]
        predict=[]
        for i in range(10): 
            next_sequence=session.run(prediction,feed_dict={X:[previous_sequence]})
            predict.append(next_sequence[-1])
            previous_sequence=np.vstack((previous_sequence[1:],next_sequence[-1]))
        plt.figure()
        plt.plot(list(range(len(normalize_data))), normalize_data, color='black')
        plt.plot(list(range(len(normalize_data), len(normalize_data) + len(predict))), predict, color='red')
        plt.show()
```

```python
# create data forecast
list_of_issues_dict_data = [json.loads(line) for line in open('SPM587SP19issues.json')]
issues_df = DataFrame(list_of_issues_dict_data)
wrangled_issues_df = issues_df[['Author','State','closed_at','created_at','issue_number','labels']]
labels = wrangled_issues_df.groupby(['created_at'])["Author"].count().reset_index()
labels.rename(columns={"Author":"y",'created_at':'ds'},inplace=True)
data =np.array(labels.y)
normalize_data=(data-np.mean(data))/np.std(data)
normalize_data=normalize_data[:,np.newaxis]
```

```python
build_data(normalize_data)
train_and_predict()
```
